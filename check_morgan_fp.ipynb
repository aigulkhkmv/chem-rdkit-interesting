```python
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
from rdkit import DataStructs
```

```python
from rdkit.Chem.Draw import SimilarityMaps
mol = AllChem.MolFromSmiles('C1=CC=NC=C1')
refmol = AllChem.MolFromSmiles('[Cl-].C1=CC=[NH+]C=C1')
```

```python
fig, maxweight = SimilarityMaps.GetSimilarityMapForFingerprint(refmol, mol, 
                                                               lambda m, idx: SimilarityMaps.GetMorganFingerprint(m, atomId=idx, radius=4, nBits=4096, useBondTypes=True, useFeatures=True),  metric=DataStructs.TanimotoSimilarity)
```

# salt fp vs not not salt

```python
from rdkit.Chem.Draw import DrawMorganBit, DrawRDKitBit
```

```python
pyridines = {
    "pyridine": "C1=CC=NC=C1",
    "pyridinium_chloride": "[Cl-].C1=CC=[NH+]C=C1",
     "pyridinium_bromide": "[Br-].C1=CC=[NH+]C=C1",
     "pyridine_charge":"Cl[N+]1=CC=CC=C1"
}
```

```python
for pyridine_name, pyridine_smi in pyridines.items():
    print(pyridine_name, pyridine_smi)
```

```python
fp_info = {}
for pyridine_name, pyridine_smi in pyridines.items():
    mol = AllChem.MolFromSmiles(pyridine_smi)
    bi = {}
    fp = AllChem.GetMorganFingerprintAsBitVect(
        mol,
        radius=4,
        nBits=4096,
        useBondTypes=True,
        useFeatures=True,
        bitInfo=bi)
    fp_info.update({pyridine_name: {"bits": fp.ToBitString(), "info": bi}})
```

```python
first = []
second = []
eq = []
for i in fp_info:
    for j in fp_info:
        first.append(i)
        second.append(j)
        eq.append(fp_info[i]["bits"] == fp_info[j]["bits"])
```

```python
import pandas as pd
pd.DataFrame({"first": first, "second": second, "bits": eq})
```

```python
from pathlib import Path
cur_path = Path("/Users/aigul/Desktop/work/check-morgan-fp/bit_info")
for name, info in fp_info.items():
    bit_info = info["info"]
    print(name, len(bit_info))
    for bit in bit_info:
        mfp2_svg = DrawMorganBit(AllChem.MolFromSmiles(pyridines[name]) , bit, bit_info, useSVG=True)
        svg = open(cur_path/f'{name}_{bit}.svg', 'w')
        svg.write(mfp2_svg)
        svg.close()
```
